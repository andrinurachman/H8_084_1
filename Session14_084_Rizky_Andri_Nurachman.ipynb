```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
#analisis data set yang telah dipanggil 
#data set array
breast = load_breast_cancer()
#define features dari data set
breast_data = breast.data
#define label dari data set
breast_label = breast.target
#lihat dimensi shape
print(breast_data.shape)
print(breast_label.shape)
```

```python
labels = np.reshape(breast_label, (569,1))
```

```python
final_breast_data = np.concatenate([breast_data,labels], axis= 1)
```

```python
final_breast_data.shape
```

```python
#buat array jadi data frame
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
breast_dataset.head(5)
```

```python
breast_dataset.tail(5)
```

```python
#ngeliat data ke 30 ini yg merupakan label digabung
breast_dataset.loc[breast_dataset[30] == 0]
```

```python
features = breast.feature_names
```

```python
features
```

```python
breast_dataset
```

```python
feature_labels = np.append(features, 'labels')
```

```python
breast_dataset.columns = feature_labels
```

```python
breast_dataset.head()
```

Mengubah nilai didalam kolom label dari 0 dan 1 jadi string benign atau malignant

```python
breast_dataset.labels.replace(0,'Benign', inplace=True)
breast_dataset.labels.replace(1,'Malignant', inplace=True)
```

```python
breast_dataset.tail()
```

```python
breast_dataset.info()
```

```python
#import classifier dari sklearn
from sklearn.preprocessing import StandardScaler
#x adalah nilai dari semua kolom yang dipanggil
x= breast_dataset.loc[:, features].values
#transform si x , normalisasi features
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
#kasih judul kolom dengan perulangan sampai maksimum 30
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
#sampai sini, normalised breast adalah data frame yang sudah dinormalisasi dengan judul kolom yang diberi angka
```

```python
#gunakan fungsi PCA
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(normalised_breast)
```

```python
#buat dataframe dengan nilai principal component untuk semua sample
principal_breast_Df = pd.DataFrame(data=principalComponents_breast,
                                  columns=['principal component 1', 'principal component 2']
                                  )
```

```python
principal_breast_Df.tail()
```

```python
principal_breast_Df.shape
```

```python
#variasi PCA
print('Variasi per Principal Component :{}'.format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['labels'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})

plt.show()
```

# K MEANS WITH PYTHON

```python
from sklearn.cluster import KMeans
```

```python
X = -2 * np.random.rand(100,2)
X1 = 1+2 * np.random.rand(50,2)
print(X.shape)
print(X1.shape)
X[50:100, :] = X1
print(X.shape)
print(X1.shape)
plt.scatter(X[:,0], X[:,1], s=50,c='b')
plt.show()
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('Wholesale customers data.csv')
data.head()
```

```python
data.describe()
```

```python
#data channel, region kecil dibandingkan yang lain, perlu standarisasi
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

#statistic dari scaled data
pd.DataFrame(data_scaled).describe()
```

```python
#define Kmeans function dengan memulai sebagai k-means++
kmeans = KMeans(n_clusters=2, init='k-means++')

#ngepasin si algoritma k means dengan scaled data (data_scaled)
kmeans.fit(data_scaled)
```

```python
#inersia
kmeans.inertia_
```

```python
# fitting multiple k-means algorithms and storing the values in an empty list
SSE = []
for cluster in range(1,20):
 kmeans = KMeans(n_clusters = cluster, init='k-means++')
 kmeans.fit(data_scaled)
 SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
# k means using 5 clusters and k-means++ initialization
kmeans = KMeans(n_clusters = 6, init='k-means++')
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)
```

```python
frame = pd.DataFrame(data_scaled)
frame['cluster'] = pred
frame['cluster'].value_counts()
```

```python

# Assigning features and label variables
weather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny', 'Rainy','Sunny','Overcast','Overcast','Rainy']
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

# Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
weather_encoded=le.fit_transform(weather)

# Converting string labels into numbers
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)

#Combinig weather and temp into single listof tuples
features=list(zip(weather_encoded,temp_encoded))

#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features,label)

# save the model to disk
import pickle

filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```
